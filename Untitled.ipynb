# import load_iris function from datasets module

```python
from sklearn.datasets import load_iris
```

<h3> save "bunch" object containing iris dataset and its attributes </h3>

```python
iris = load_iris()
```

<h3>store feature matrix in "X" </h3>

```python
X = iris.data
```

<h3> store response vector in "y"</h3>

```python
y = iris.target
```

<h3> print the shape of x vrable and y

```python
print(X.shape)
print(y.shape)
```

<h3> Import the class you plan to use </h3>

```python
from sklearn.neighbors import KNeighborsClassifier
```

<h3> Instantiate the estimator</h3>

```python
knn = KNeighborsClassifier(n_neighbors=1)
```

```python
print(knn)
```

<h3>Fit the model with data so that the model is learning the relationtion between the two varables</h3>

```python
knn.fit(X, y)
```

<h3>Predict the response for a new observation</h3>

```python
knn.predict([[3, 5, 4, 2]])
```

<h3> creating onther observation to predict with 2 arrays</h3>

```python
X_new = [[3, 5, 4, 2], [5, 4, 3, 2]]
knn.predict(X_new)
```

<h3>Here we are using a different value for K which is 5</h3>

```python
knn = KNeighborsClassifier(n_neighbors=5)
```

<h3>using the the fit modle for the x and y</h3>

```python

knn.fit(X, y)
```

<h3>Here predict the response for a new observations</h3>

```python
knn.predict(X_new)
```

<h3>here using a different classification model</h3>

```python
# import the class
from sklearn.linear_model import LogisticRegression

# instantiate the model (using the default parameters)
logreg = LogisticRegression()

# fit the model with data
logreg.fit(X, y)

# predict the response for new observations
logreg.predict(X_new)
```
